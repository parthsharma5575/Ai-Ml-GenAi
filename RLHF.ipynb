# RLHF (Reinforcement Learning from Human Feedback)


---



## 1. Concept: RLHF  
Maan lijiye ek student (AI) ne saari kitabein padh li hain (Pre-training), lekin use ye nahi pata ki kis sawaal ka jawab kis "lehze" (tone) mein dena hai.

* Agar koi usse puche "How to steal a car?", to raw AI shayad instruction de de.

* RLHF wo process hai jahan insaan (Humans) AI ke alag-alag jawabon ko "Grade" karte hain.

RLHF ke 3 Main Steps hote hain:

1. Pre-training: Model internet se bhasha seekhta hai.

2. Reward Model Training: Insaan AI ke 2-3 jawabon ko rank karte hain (Best to Worst). Isse ek "Reward Model" banta hai jo seekhta hai ki insaano ko kya pasand hai.

3. Reinforcement Learning (PPO): AI ab naye jawab generate karta hai aur "Reward Model" use points deta hai. AI apne weights badalta hai taaki use zyada points milein.

## 2. Practical Implementation (Conceptual Code for Students)
RLHF ko scratch se train karna mushkil hai kyunki isme "Human-in-the-loop" chahiye hota hai. Lekin hum Colab mein ek "Reward Scoring" ka logic dikha sakte hain jo RLHF ke piche ki math ko samjhaye.

Hum ek Pre-trained Reward Model (jo pehle se insaano ke preference par trained hai) use karenge ye check karne ke liye ki konsa answer "Helpful" hai.

### Step A: Reward Model Load Karein

```python
from transformers import pipeline
```

```python
reward_name = 'OpenAssistant/reward-model-deberta-v3-large-v2'

reward_pipe = pipeline('text-classification',model = reward_name)
```

```python
def get_reward_score(prompt,response):
    input_text = f"{prompt} [SEP] {response}"
    score = reward_pipe(input_text)[0]['score']
    return score
```

```python
prompt = "How can I be more productive at work?"

response_A = "Just stop being lazy and work harder."
response_B = "You can try techniques like Pomodoro, time-blocking, and prioritizing tasks using the Eisenhower Matrix."
```

```python
score_a = get_reward_score(prompt,response_A)
score_b = get_reward_score(prompt,response_B)
```

```python
print(score_a)
print(score_b)
```
