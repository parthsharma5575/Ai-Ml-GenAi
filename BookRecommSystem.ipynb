Merge Columns

```python
books=df[['title','authors','subtitle','categories','description','average_rating','published_year']]
```

```python
books.head()
```

```python
books=books.copy()
books['content']=books['title']+' '+books['subtitle']+' '+books['authors']+' '+books['categories']+' '+books['description']
```

Vectorize and cosine_similarity

```python
tfid=TfidfVectorizer(stop_words='english',max_features=5000)
```

```python
tfid_matrix=tfid.fit_transform(books['content'])
```

```python
similarity=cosine_similarity(tfid_matrix)
```

```python
print(similarity)
```

Recommendation Function

```python
from difflib import get_close_matches

def recommend(bookname,n=10):
  bookname=bookname.strip()
  idx=books[books['title'].str.lower()==bookname.lower()].index
  #exact match
  if len(idx)>0:
    print("Exact match found :\n ")
    display(books.loc[idx][['title','authors','categories']])
    idx = idx[0]
  #if not exact match
  else:
    closest=get_close_matches(bookname,books['title'].tolist(),n=1,cutoff=0.6)
    if len(closest)>0:
      print(f"Book not found,")
      print(f"Showing results for '{closest[0]}'\n")
      idx=books[books['title']==closest[0]].index[0]
      display(books.loc[[idx]][['title','authors','categories']])
    else:
      #author search
      author=books[books['authors'].str.contains(bookname,case=False,na=False)]
      if len(author)>0:
        print("Author matched.\n")
        display(author[['title','authors']].head(10))
        idx=author.index[0]
      else:
        #desc search
        desc=books[books['description'].str.contains(bookname,case=False,na=False)]
        if len(desc)>0:
          print("Description matched.\n")
          display(desc[['title','authors']].head(10))
          idx=desc.index[0]
        else:
          return "No Book Found.\n"
  recomm=[]

  distances=list(enumerate(similarity[idx]))
  distances=sorted(distances,key=lambda x:x[1],reverse=True)


  for i in distances[1:n+1]:
    recomm.append([books.iloc[i[0]].title,
                  books.iloc[i[0]].authors,
                  books.iloc[i[0]].categories,
                  books.iloc[i[0]].description]
                  )
  print("Recommended Books\n")
  return pd.DataFrame(recomm,columns=['Title','Authors','Categories','Description'])
```
