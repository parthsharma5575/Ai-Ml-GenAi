### **Remove html tags**

```python
text = "<html><body><p> Movie 1</p><p> Actor - Aamir Khan</p><p> Click here to <a href='http://google.com'>download</a></p></body></html>"
```

```python
import re
```

```python
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'', text)
```

```python
remove_html_tags(text)
```

```python
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'', text)
```

### **Chat word correcter**

```python
chat_words = {'IMHO':'In My Honest/Humble Opinion', 'GN':'Good night','GM':'Good Morning'}
```

```python
def actual(sentence):

  chat_words = {'IMHO':'In My Honest/Humble Opinion', 'GN':'Good night','GM':'Good Morning'}

  for i in sentence:
    if i in string.punctuation:
      sentence = sentence.replace(i,'')


  for i in sentence.split():
    if i.upper() in chat_words:
      sentence = sentence.replace(i,chat_words[i.upper()])

  return sentence
```

```python
actual('gm how are you')
```

**Remove emoji**

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
remove_emoji("Loved the movie.😘😘 It was 🔥")
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return ' '.join([ps.stem(word) for word in text.split()])
```

## Tokens :-
>Tokenization ka matlab hai kisi bade sentence ya document ko chhote-chhote tukdon (units) mein todna. In tukdon ko hum Tokens kehte hain.

### 1. Type of Tokenization :-

1.  Word-level Tokenization:-

Har word ek token hota hai.

    Sentence: "AI is amazing"

    Tokens: ["AI", "is", "amazing"]

    Problem: Agar "amazing" ki jagah "amazingly" aa jaye, to AI confuse ho sakta hai agar usne wo word pehle nahi dekha. Isse vocabulary bahut badi ho jati hai.

2. Character-level Tokenization :-

Har letter ek token hota hai.

    Sentence: "AI"

    Tokens: ["A", "I"]

    Problem: Isse sentence bahut lamba ho jata hai aur model ko "matlab" (meaning) nikalne mein mushkil hoti hai.

3. Subword Tokenization :-

(Jo modern AI use karta hai - e.g., GPT, BERT)
Ye words ko aise todta hai ki common hisse (prefixes/suffixes) alag ho jayein.

    Example Word: "Unbelievable"

    Tokens: ["Un", "believ", "able"]

    Benefit: Isse AI naye words ka matlab bhi samajh sakta hai unke tukdon ko dekh kar.
