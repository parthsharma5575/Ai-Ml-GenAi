## 1. Encoder


---



Encoder ka kaam hai input text ko poori tarah se "padhna" aur uska gehra matlab (context) nikalna.

* Kaise kaam karta hai? Ye Bi-directional hota hai. Matlab, agar ek sentence hai "The bank is closed," to Encoder "bank" word ko samajhne ke liye uske piche wale words aur aage wale words dono ko ek saath dekhta hai.

* Kyun use hota hai? Taaki model sentence ka "Nichen" (Summary/Intent) samajh sake.

* BERT (Bidirectional Encoder Representations from Transformers): Iska main kaam hai Classification, Translation (input side), aur Sentiment Analysis.

Example:-

ek **"Student"** hai jo exam mein poora paragraph padh kar uska answer dhoondh raha hai. Wo upar-neeche kahin bhi dekh sakta hai context ke liye.

## 2. Decoder


---



GPT Decoder ka kaam hai ek-ek karke naye words "paida" (generate) karna.

* Kaise kaam karta hai? Ye Uni-directional (Auto-regressive) hota hai. Matlab, ye sirf piche dekhe gaye words ke basis par agla word predict karta hai. Ye aage ka rasta nahi dekh sakta.

* Masked Self-Attention: Decoder mein hum future words ko "chhupa" (mask) dete hain taaki model cheating na kare aur khud se predict kare.

* GPT (Generative Pre-trained Transformer): Iska main kaam hai Content likhna, Chatting karna, aur Stories banana.

Example :-

ek "Author" hai jo ek naya novel likh raha hai. Wo piche likhe pannon (pages) ko to janta hai, lekin aage kya hone wala hai wo usse khud likhna (predict) hai.

### Encoder (BERT) Implimentation

```python
from transformers import pipeline
```

```python
nlp_encoder = pipeline('fill-mask',model = 'bert-base-uncased')
```

```python
text = 'The capital of France is [MASK].'
```

```python
result = nlp_encoder(text)
print(result)
```

```python
for i,j in enumerate(result):
  print(f'pred {i+1}: {j['token_str']} (confidance : {j['score']:.4f})')
```

### Decoder (GPT) Implimentation

```python
from transformers import pipeline,set_seed
```

```python
genrator = pipeline('text-generation',model = 'gpt2')

set_seed(42)
```

```python
prompt = 'The Future of Artificial intelligence is'
```

```python
result = genrator(prompt,max_length = 30,num_return_sequences = 1)
print(result[0]['generated_text'])
```

## Parameters


---



### 1. The Analogy: The Giant Mixing Board

socho ek bahut bada Music Mixing Board hai jisme 175 billion buttons aur sliders lage hue hain.

* Har slider ek "Weight" hai.

* Agar aap ek slider ko thoda upar karte hain, to "Bass" badh jata hai. Agar dusre ko niche karte hain, to "Vocals" saaf ho jate hain.

* AI training ka matlab hai in 175 billion sliders ko sahi jagah par set karna, taaki jab aap "Hello" kahein, to output hamesha "How can I help you?" hi aaye.

2. "Weights" Asal Mein Kya Hain?

Technical terms mein, Parameters = Weights + Biases.
* Weights ($W$): Ye batate hain ki kisi specific information ki "importance" kitni hai.
  * Example: Agar model agla word predict kar raha hai, to pichle sentence mein "Weather" word ki weight zyada hogi agar use "Rain" predict karna hai.
  * The Connection: LLM ek bahut bada "Neural Network" hai. Har neuron doosre neuron se juda hota hai. In connections ki "strength" ko hi hum Weight kehte hain hain.

### 3. 175 Billion Numbers ka Size
Ek parameter asal mein ek Decimal Number hota hai (jaise 0.0054 ya -1.22).
* Agar hum ek number ko store karne ke liye 2 bytes (FP16 precision) use karein:
* $175,000,000,000 \times 2 \text{ bytes} \approx 350 \text{ GB}$.
* Matlab, sirf model ko "load" karne ke liye aapko 350 GB ki High-speed VRAM (GPU memory) chahiye. Isi liye bade models ko chalane ke liye mehnge GPUs (A100/H100) ki zaroorat padti hai.

### 4. Practical Example: Predict the Next Number
Maan lijiye hamara model hai: $y = w \cdot x$

* $x$ = Input (Token ID)
* $w$ = Weight (Parameter)
* $y$ = Prediction
* $x = 10$ aur model ko answer $20$ lana hai, to training ke time AI $w$ ko 2 par set kar dega ($10 \times 2 = 20$).LLMs mein yahi cheez 175 billion baar hoti hai, jahan alag-alag layers mil kar complex patterns  seekhti hain.

### 5. Why more is (usually) better?

* Low Parameters: Model sirf simple patterns seekh pata hai (jaise sirf spelling check).

* High Parameters (175B+): Model "Reasoning" aur "World Knowledge" seekh leta hai. Use pata chal jata hai ki "Newton" ka relation "Apple" aur "Gravity" dono se hai.
* Parameters model ki "Memory" aur "Intelligence" ka size hain. Jitne zyada parameters, utni hi bariki se model language ke patterns ko samajh sakta hai.
