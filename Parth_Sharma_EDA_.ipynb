## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df_new=df.copy()
````

````python
# Write your code to make your dataset analysis ready.
df_new.isnull().sum()
````

````python
df_new.info()
````

````python
#No Club
df_new['Club']=df_new['Club'].fillna('No Club')
````

````python
df_new[df_new['Value'].isnull()][['Name','Club','Position','Wage']]
````

````python
df_new[df_new['Value'].isnull()][['Name','Club','Wage']]
#value is null even for players with wage and club present so cannot set null to 0 for value
````

````python
df_new[df_new['Wage'].isnull()][['Name','Club']]
````

````python
#Value 0 is not ideal so filling with median
df_new['Value']=df_new['Value'].fillna(df_new['Value'].median())
````

````python
df_new['International Reputation']=df_new['International Reputation'].fillna(df_new['International Reputation'].mode()[0])
````

````python
df_new['Skill Moves']=df_new['Skill Moves'].fillna(df_new['Skill Moves'].mode()[0])
````

````python
df_new['Contract Valid Until']=pd.to_datetime(df_new['Contract Valid Until'])
````

````python
df_new.dtypes
````

````python
df_new.isnull().sum()
````

### What all manipulations have you done and insights you found?

I created a working copy of the original dataset and focused on handling missing values and correcting data types so that the data is reliable for analysis. For categorical fields, I filled missing Club values with “No Club” to clearly mark unattached players, and replaced missing International Reputation and Skill Moves with their most frequent (mode) values to keep distributions realistic. For numerical fields, I imputed missing Value using the median instead of setting it to zero, which avoids artificially undervaluing players. I also converted Contract Valid Until to a proper datetime type to support any time‑based analysis on contracts. After these steps, the dataset has no missing values except a few contracts that still have null end dates, highlighting players with uncertain or non‑standard contract information.

#### Chart - 1

````python
# Top 10 Clubs by Total Market Value
club_val=df_new.groupby('Club')['Value'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=club_val.values,y=club_val.index)
plt.title('Top 10 clubs by Total Market Value')
plt.xlabel('Total Value')
````

##### 1. Why did you pick the specific chart?

I chose a horizontal bar chart because it clearly ranks clubs by total squad market value and makes it easy to compare the top 10 clubs side by side.

##### 2. What is/are the insight(s) found from the chart?

The chart shows that Real Madrid has the most valuable squad, followed closely by FC Barcelona and a few other elite European clubs, indicating a strong concentration of player value in top‑tier teams.

#### Chart - 3

````python
# Average wage by position
avg_wage=df_new.groupby('Position')['Wage'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.title('Average Wage by Position')
plt.xlabel('Position')
plt.ylabel('Average Wage')
sns.barplot(x=avg_wage.index,y=avg_wage.values)
plt.xticks(rotation=90)
plt.show()
````

##### 1. Why did you pick the specific chart?

I used a bar chart because it clearly compares average wages across different positions and highlights which roles are paid the most.

##### 2. What is/are the insight(s) found from the chart?

The chart shows that attacking positions like strikers and wide forwards tend to have higher average wages than many defensive or midfield roles, reflecting their premium value in the market.

##### 3. Will the gained insights help creating a positive business impact?
