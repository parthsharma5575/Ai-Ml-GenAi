## Context Window
LLM ki wo "Short-term Memory" hai jo tai karti hai ki model ek baar mein kitna bada data (text) process kar sakta hai.

---



## 1. Concept: Context Window Kya Hai? (The Desk Analogy)
Sochiye aap ek exam de rahe hain aur aapke paas ek choti si Desk hai.

* Aap desk par ek baar mein sirf 5 panna (pages) hi rakh sakte hain.

* Agar aapko 6th panna padhna hai, to aapko 1st panna desk se hatana padega.

* Iska matlab, 6th panna padhte waqt aap 1st panne ki baatein bhool jayenge.

AI ke saath bhi aisa hi hota hai. Context Window wo limit hai (measured in Tokens) jiske andar model saari information ko "ek saath" yaad rakh kar process kar sakta hai.

## 2. Ye Limited Kyun Hota Hai? (The Technical Reason)
Transformers mein "Self-Attention" mechanism hota hai. Iska matlab har token ko doosre har token se compare kiya jata hai.
* Agar tokens double honge, to computational power 4 guna (Quadratic $N^2$) badh jayegi.
* Zyada context matlab zyada GPU memory aur zyada paisa.

Popular Models ki Limits:

* BERT: 512 Tokens (Bahut chota)

* GPT-3.5: 4,000 - 16,000 Tokens

* GPT-4o: 128,000 Tokens (Lagbhag ek puri kitab)

* Gemini 1.5 Pro: 1 Million+ Tokens

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('gpt2')
```

```python
max_limit = tokenizer.model_max_length

print(max_limit)
```

```python
long_text = 'this is a sentance. '*600

tokens_all = tokenizer.encode(long_text)

print('totla len :-',len(tokens_all))
```

```python
tokens_all
```

```python
token_truncarted  = tokenizer.encode(
    long_text,
    max_length = max_limit,
    truncation = True

)
```

```python
print('tokens after truncate:-',len(token_truncarted))
```

```python
print(len(tokens_all) - len(token_truncarted))
```

## 4. Points highlight?
1. Forgetting in Conversation: Agar aap AI se bahut lambi baat karenge, to wo conversation ke shuruati instructions bhool sakta hai agar context window bhar gayi ho.

2. Tokens vs. Words: 1000 words lagbhag 1300-1400 tokens hote hain. Limit hamesha tokens par hoti hai, words par nahi.

3. The "Lost in the Middle" Problem: Research kehti hai ki bade context window mein bhi models aksar sentence ke beech ki information bhool jate hain aur shuruat ya aakhir par zyada dhyan dete hain.
